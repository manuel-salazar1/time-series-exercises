#### 1. Attendance Data

Read the data from the attendance table and calculate an attendance percentage for each student. One half day is worth 50% of a full day, and 10 tardies is equal to one absence.

```python
dfa = pd.read_csv(attendance)
```

```python
dfa
```

```python
dfa.rename(columns={'Unnamed: 0': 'name'}, inplace=True)
```

```python
dfa
```

```python
dfa_melt = dfa.melt(id_vars='name',
        var_name='date',
        value_name='attendance')
```

```python
dfa_melt['point'] = 0

for i, row in dfa_melt.iterrows():
    if row['attendance'] == 'A':
        dfa_melt.loc[i, 'point'] = -1
    else:
        if row['attendance'] == 'H':
            dfa_melt.loc[i, 'point'] = -0.5
        elif row['attendance'] =='T':
            dfa_melt.loc[i, 'point'] = -0.1
```

```python
dfa_melt.head()
```

```python
100 - ((dfa_melt.groupby('name').point.sum() / 8) * -100)
```

#### 3. Cake Recipes

a. Read the cake_recipes table. This data set contains cake tastiness scores for combinations of different recipes, oven rack positions, and oven temperatures.  
b. Tidy the data as necessary.  
c. Which recipe, on average, is the best?  
d. Which oven temperature, on average, produces the best results?  
e. Which combination of recipe, rack position, and temperature gives the best result?

```python
dfk = pd.read_csv(cake)
```

```python
dfk
```

```python
dfk[['recipe', 'position']]= dfk['recipe:position'].str.split(':', expand=True)
```

```python
dfk
```

```python
dfk.drop(columns = 'recipe:position', inplace = True)
```

```python
dfk
```

```python
max_score = 0
max_recipe = None
max_position = None

for i, row in dfk.iterrows():
    if max(row['225'], row['250'], row['275'], row['300']) > max_score:
        max_score = max(row['225'], row['250'], row['275'], row['300'])
        max_recipe = row['recipe']
        max_position = row['position']
        
print(f'Highest score: ', round(max_score, 2))
print(f'Recipe: ', max_recipe)
print(f'Position: ', max_position)
```

```python
dfk['225'].mean(), dfk['250'].mean(), dfk['275'].mean(), dfk['300'].mean()
```

```python
best_ave_score = 0
best_recipe = None
best_position = None

for i, row in dfk.iterrows():
    if (row['225'] + row['250'] + row['275'] + row['300']) > max_score:
        max_score = (row['225'] + row['250'] + row['275'] + row['300'])
        max_recipe = row['recipe']
        max_position = row['position']
        
print(f'Highest score: ', round(best_ave_score, 2))
print(f'Recipe: ', best_recipe)
print(f'Position: ', best_position)
```
